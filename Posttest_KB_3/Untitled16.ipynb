```python
x = df.Brand_Name[0:5]
y = df.How_Many_Sold[0:5]
plt.figure(figsize = (5, 3))
plt.title("Sebuah Grafik")
plt.xlabel("Brand_Nama")
plt.ylabel("How_Man_Sold")
plt.plot(x, label = "Brand_Nama", color = "pink")
plt.plot(y, label = "How_Man_Sold", color ="purple",marker = "o")
plt.legend(loc = "upper right")
plt.show()
```

```python
Diagram = df['Current_Price'].value_counts().rename_axis('Current_Price').reset_index(name='Id')
plt.figure(figsize=(20, 20))
plt.pie(Diagram.Id,labels = Diagram.Current_Price, startangle=90, autopct="%.1f%%")
plt.legend()
```

```python
x = list(df.RATING[0:5])
y = list(df.RATING[0:5])
plt.bar(x,y, color="pink")
plt.xlabel("Nilai Rating X")
plt.ylabel("Nilai Rating Y")
plt.title("Perbandingan Rating")
plt.show()
```
